## Normalize the data

The data need to be normalized for a ternary plot. We'll normalize to 100:

```python
cols = ['Quartz', 'K-Feldspar', 'Plagioclase']

for col in cols:
    df_modal_abundance[col[0]] = df_modal_abundance[col] * 100 / df_modal_abundance[cols].sum(axis=1)
```

```python
df_modal_abundance.head()
```

```python
import ternary
```

```python
# Set up the figure.
fig, tax = ternary.figure(scale=100)
fig.set_size_inches(14, 12)

# Plot points.
tax.scatter(df_modal_abundance[['P', 'Q', 'K']].values)

# Axis labels. (See below for corner labels.)
fontsize = 14
offset = 0.08
tax.left_axis_label("Feldspars %", fontsize=fontsize, offset=offset)
tax.right_axis_label("Quartz %", fontsize=fontsize, offset=offset)
tax.bottom_axis_label("Plagioclase %", fontsize=fontsize, offset=-offset)
tax.set_title("Sedimentary Formations", fontsize=20)

# Corner labels.
fontsize = 12
offset = 0.1
tax.top_corner_label("Q", fontsize=fontsize, offset=0.2)
tax.left_corner_label("K", fontsize=fontsize, offset=offset)
tax.right_corner_label("P", fontsize=fontsize, offset=offset)

# Decoration.
tax.boundary(linewidth=1)
tax.gridlines(multiple=10, color="gray")
tax.ticks(axis='lbr', linewidth=1, multiple=20)
tax.get_axes().axis('off')

tax.show()
```

```python
# Set up the figure.
fig, tax = ternary.figure(scale=100)
fig.set_size_inches(10, 9)



# Plot the points in groups.
for name, group in df_modal_abundance.groupby('Stream Section'):
    
    # Note that we have to shuffle the order.
    # This will place Q at the top, F on the left.
    # So the column order is: right, top, left.
    points = group[['P', 'Q', 'K']].values
    tax.scatter(points, marker='o', label=name)
    
# Axis labels. (See below for corner labels.)
fontsize = 14
offset = 0.08
tax.left_axis_label("Feldspars %", fontsize=fontsize, offset=offset)
tax.right_axis_label("Quartz %", fontsize=fontsize, offset=offset)
tax.bottom_axis_label("Plagioclase %", fontsize=fontsize, offset=-offset)
tax.set_title("Sedimentary Formations", fontsize=20)

# Corner labels.
fontsize = 20
offset = 0.1
tax.top_corner_label("Q", fontsize=fontsize, offset=0.2)
tax.left_corner_label("K", fontsize=fontsize, offset=offset)
tax.right_corner_label("P", fontsize=fontsize, offset=offset)

# Decoration.
tax.legend(title="Stream Section")
tax.boundary(linewidth=1)
tax.gridlines(multiple=10, color="gray")
tax.ticks(axis='lbr', linewidth=1, multiple=20)
tax.get_axes().axis('off')

tax.show()
```

```python
df_geochemistry= pd.read_excel (r'C:\Users\698422\Desktop\Teaching\Teaching\Honours Class\Sedimentary Data\Garzanti sedimentary data.xlsx', sheet_name='Geochemistry Data')
print (df_geochemistry)
```

```python
fig,ax = plt.subplots()
sns.scatterplot(data=df_geochemistry, hue='Sample', x='Ni', y='TiO2')
#place legend outside top right corner of plot
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0, title='Sample')
plt.title('Binary plot of Ni vs. TiO2 showing the provenance of sediment')
plt.xlabel('Ni (ppm)')
plt.ylabel('TiO2 (wt%)')
plt.savefig('scatter.png')
plt.show()
```

```python
sum_column = df_geochemistry["Na2O"] + df_geochemistry["CaO"]
df_geochemistry["Na2O + CaO"] = sum_column
print(df_geochemistry)
df_geochemistry.head()
```

```python
cols = ['Na2O + CaO', 'Al2O3', 'K2O']

for col in cols:
    df_geochemistry[col[0]] = df_geochemistry[col] * 100 / df_geochemistry[cols].sum(axis=1)
```

```python
df_geochemistry.head()
```

```python
# Set up the figure.
fig, tax = ternary.figure(scale=100)
fig.set_size_inches(10, 9)



# Plot the points in groups.
for name, group in df_geochemistry.groupby('Sample'):
    
    # Note that we have to shuffle the order.
    # This will place Q at the top, F on the left.
    # So the column order is: right, top, left.
    points = group[['K', 'A', 'N']].values
    tax.scatter(points, marker='o', label=name)
    
# Corner labels.
fontsize = 20
offset = 0.1
tax.top_corner_label("Al2O3 %", fontsize=fontsize, offset=0.2)
tax.left_corner_label("(Na2O + CaO*) %", fontsize=fontsize, offset=offset)
tax.right_corner_label("K2O %", fontsize=fontsize, offset=offset)

# Decoration.
tax.legend(title="Sample")
tax.boundary(linewidth=1)
tax.gridlines(multiple=10, color="gray")
tax.ticks(axis='lbr', linewidth=1, multiple=20)
tax.get_axes().axis('off')

tax.show()
```
